## Perceptron Learning algorithm

### Obtain the coordinates of a point and classification label

```python
X = np.mat(data.iloc[:,:4])
y = np.mat(data.iloc[:,-2])
```

### Classify the point based on the weight

```python
def classify(weight, X):
    return np.mat(np.where(X * weight.T >= 0, 1, -1))
```

### Updata the weight each iteration

```python
def updata_weight(weight, X, y, a):  ## the parameter "a" determine the learning rate.
    diff = y.T - classify(weight, X)
    weight += diff.T/a * X   # when a is larger, the learning rate is slower.
    return weight
```

### Calculate the accuracy

```python
def get_accuracy(weight, X, y):
    m = X.shape[0]
    diff = y.T - classify(weight, X)
    misclassify_point = np.sum(diff != 0)
    return 1 - misclassify_point/m
```

### Train the model

```python
def fit_PLA(X, y):
    weight = np.mat(np.zeros(X.shape[1]))
    separate = False
    count = 0
    while (True):
        count += 1
        weight = updata_weight(weight, X, y, 2000)
        if(get_accuracy(weight, X, y) == 1):
            break
    
    return weight
```

```python
weight = fit_PLA(X, y)
weight
```

```python
get_accuracy(weight, X, y)
```

## Pocket algorithm

### Obtain the coordinates of a point and classification label

```python
X = np.mat(data.iloc[:,:4])
y = np.mat(data.iloc[:,-1])
```

### Train the pocket model

```python
def fit_pocket(X, y):  #This algorithm keeps the best result seen so far. So I write it based on the PLA model
    weight = np.mat(np.zeros(X.shape[1]))
    m = X.shape[0]
    count = 0
    max_accuracy = 0
    misclassified_list = list()
    best_weight = None;
    while (count < 7000):
        count += 1
        weight = updata_weight(weight, X, y, 2000)
        current_accuracy = get_accuracy(weight, X, y)
        if (max_accuracy < current_accuracy):
            max_accuracy = current_accuracy
            best_weight = weight.copy()
            
        misclassified_list.append((1 - current_accuracy)*m)
        if(get_accuracy(weight, X, y) == 1):
            break
        
    return best_weight, misclassified_list
```

```python
weight, misclassified_list = fit_pocket(X, y)
weight
```

```python
get_accuracy(weight, X, y)
```

### Plot the number of misclassified points against the number of iterations of the algorithm.

```python
def plot_missclass(missMat):
    plt.figure(figsize=(20,8))
    x_axis = []
    y_axis = []
    for i in range(7000):
        if i % 19 == 0:
            x_axis.append(i)
            y_axis.append(missMat[i])
        
    plt.plot(x_axis, y_axis)
    plt.xlabel('Iterations')
    plt.ylabel('Misclassifications')
    plt.title('The Number of Misclassified points')
    plt.show()
```

```python
plot_missclass(misclassified_list)
```

## Logistic Regression

### Obtain the coordinates of a point and classification label

```python
X = np.mat(data.iloc[:,:4])
y = np.mat(data.iloc[:,-1])
X, y
```

### Sigmoid function

```python
def sigmoid(z):
    return (1/(1 + np.exp(-z)))
```

$E_{in}(w)$

```python
def gradient(weight, X, y):
    m = X.shape[0]
    Ein = np.multiply(1 / (1 + np.exp(np.multiply(y.T, (weight * X.T).T).T)), y) * X
    return -1/float(m)*Ein
```

### Train the Logistic Regression model

```python
def fit_LG(X, y, a):
    count = 0
    weight = np.mat(np.zeros(X.shape[1]))
    
    while (count < 7000):
        weight -= a * gradient(weight, X, y)
        count += 1
    return weight
```

```python
weight = fit_LG(X, y, 1)
weight
```

## Linear Regression

### Obtain original data

```python
data = pd.read_csv("linear-regression.txt", header = None, names = ["X", "Y", "Z"])
data.insert(0, 'X0', 1)
data.head()
```

```python
X = np.mat(data.iloc[:, :3])
y = np.mat(data.iloc[:, -1])
```

### Train the Linear Regression model

```python
def fit_linear_regression(X, y):
    D = X.T
    weight = np.linalg.inv(D * D.T) * D * y.T
    
    return weight.T
```

```python
weight = fit_linear_regression(X, y)
weight
```
